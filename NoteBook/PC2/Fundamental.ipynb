```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(0, 4 * np.pi, 100)
x=theta
y=np.cos(x)
z=-np.sin(x)
ax.plot(x, y, z, label='(u(t),v(t))')
plt.title("Integer curve")
ax.legend()

plt.show()
```

## 2.2.1 curve and orbit

```python
t=np.linspace(0,np.pi*10 ,1000)
fig = figure(plot_height=225, plot_width=450, x_range=(0,np.pi*10), 
             y_axis_label = "u(t)",
             title = "Solution u(t)")
plt = fig.line(t,np.cos(t), line_width=2)
show(fig, notebook_handle=True)
t=np.linspace(0,np.pi*10 ,1000)
fig = figure(plot_height=225, plot_width=450, x_range=(0,np.pi*10), 
             y_axis_label = "v(t)",
             title = "Solution v(t)")
plt = fig.line(t,-np.sin(t), line_width=2)
show(fig, notebook_handle=True)
```

```python
t=np.linspace(0,np.pi*2)
fig = figure(plot_height=450, plot_width=450, x_range=(-1.5,1.5), y_range=(-1.5,1.5),
             x_axis_label = "u(t)",
             y_axis_label = "v(t)",title = "Phase portrait")
plt = fig.line(np.cos(t), np.sin(-t), line_width=2)
show(fig, notebook_handle=True)
```

## 2.2.2

```python
def x(t,x0):
    return t-1+np.exp(-t)*(x0+1)

x0=-1

t=np.linspace(0,50,10000)
fig = figure(plot_height=450, plot_width=900, x_range=(0,20),y_range=(0,20), 
             x_axis_label = "t",
             y_axis_label = "x(t)",title = "solution (x,t)")
plt = fig.line(t, x(t,x0), line_width=2)
plt = fig.line(t, t, line_width=2)
show(fig, notebook_handle=True)
```
